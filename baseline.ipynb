```python
from baseline import *
from data_utils import *
```

```python
#Fetch the training data
train_data = makeTrainingData(
    train_csv='train.csv',
    decription_csv='product_descriptions.csv',
    attrib_csv='attributes.csv',
    merge_col='product_uid'
)
train_data.head(6)
```

```python
#Fill Na in the brand name
train_data['brand'].fillna('',inplace=True)

# basic spell correct search query
scb = SpellingCorrectionBaseline()

train_data['search_term'] = train_data['search_term'].apply(lambda x: scb.spellCorrectBaseline(x))
scb.dicAppender()
```

```python
#Rank the prouducts based relavence
train_ranked = rankByRelevance(train_data)
train_ranked.head(5)
```

```python
#Cleaning the data and removing unnecessary columns
train_cleaned = cleanPandasDF(
    ranked_df=train_ranked,
    col_list_clean=['product_title','search_term','brand','product_description'],
    col_list_drop=['id']
)
train_cleaned.head(5)
```

```python
train_cleaned['brand'] = train_cleaned['brand'].apply(lambda x: " ".join(list(x)))
train_cleaned['product_description'] = train_cleaned['product_description'].apply(lambda x: " ".join(list(x)))
train_cleaned['product_title'] = train_cleaned['product_title'].apply(lambda x: " ".join(list(x)))
train_cleaned['search_term'] = train_cleaned['search_term'].apply(lambda x: " ".join(list(x)))
```

```python
train_cleaned
```

```python
train_cleaned.to_csv('train_dssm3.csv',header=True,index=False)
```
